# *`Séparer les variables indépendantes (features) et dépendante (target)`*

```python
X=new_df.drop(columns="Class",axis=1)
Y=new_df["Class"]
```

# *`Séparer les données d'entrainements et de validation`*

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=0)
```

# *`Standardisation des données`*

```python
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

# *`Modèles d'entrainement`*

```python
def models(X_train, Y_train):
   #Regression Logistique
   log=LogisticRegression(random_state = 0)
   log.fit(X_train, Y_train)
   #Arbre de décision
   tree = DecisionTreeClassifier(criterion = 'entropy', random_state=0)
   tree.fit(X_train, Y_train)
   #forêt d'arbres décisionnels
   forest=RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
   forest.fit(X_train,Y_train)
   # MLP Classifier (Multi-Layer Perceptron)
   mlp = MLPClassifier(random_state=0)
   mlp.fit(X_train, Y_train)
   #afficher les précisions des modèles
   return log,tree,forest,mlp
```

```python
model=models(X_train,Y_train)
```

# *`Accuracy score`*

```python
for i in range( len(model)):
   print( 'Model ', i)
   X_train_prediction=model[i].predict(X_train)
   training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
   print("Accuracy :",training_data_accuracy)
   print()
```
